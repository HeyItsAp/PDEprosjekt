### d) Utregning av fluksen
Ifølge bevaringsloven beskriver funksjonen
u_t + J_x = 0 fluksen.

I en transport sammenheng (som vi har her), er formelen for fluks gitt ved: 
J(x, t) = u(x, t) * p(x, t), der  u er tetthet og u er hastighet.
Hvis vi plugger våre definisjoner får vi at:
J(x, t) = u(x, t) * v(v(x, t)). Hastigheten vår er en funksjon av tettheten, som gjør også fluksen en hastighet av tettheten.
