```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# ------------------------------------------------------------
# Oppsett for figurer
# ------------------------------------------------------------
plt.close("all")
plt.figure(figsize=(6, 4))

# ------------------------------------------------------------
# Fluks for trafikkFlytten: u_t + (J(u))_x = 0, der J(u) = u * v(u), der v(u) = v_max(1-(u/u_max)^p)
# ------------------------------------------------------------
def flux_trafikk(u, p, u_max, v_max):
    return u*speed(u, p, u_max, v_max)

def speed(u, p, u_max, v_max):
    return v_max*(1-(1-(u/u_max))**p)


# ------------------------------------------------------------
# Lax–Friedrichs-metoden (1D, eksplisitt)
# ------------------------------------------------------------
# fluks: funksjon f(u)
# u: array med løsningen ved tid t (på romgitteret)
# dx: steglengde i rom
# dt: steglengde i tid
# t: nåværende tid (brukes av randbetingelsene)
# u_left(t): randverdi på venstre kant (x = a)
# u_right(t): randverdi på høyre kant (x = b)
def lax_friedrichs_step(fluks, u, dx, dt, t, u_left, u_right, u_max, v_max, p):
    """Utfører ett tidssteg med Lax–Friedrichs."""
    n = u.size
    u_next = np.zeros_like(u)

    # Indre punkter (i = 1, ..., n-2)
    # Formelen:
    # u_i^{n+1} = 1/2 (u_{i-1}^n + u_{i+1}^n) - (dt/(2dx)) (f(u_{i+1}^n) - f(u_{i-1}^n))
    for i in range(1, n - 1):
        u_next[i] = 0.5 * (u[i - 1] + u[i + 1]) - (dt / (2 * dx)) * (
            fluks(u[i + 1], p, u_max, v_max) - fluks(u[i - 1], p, u_max, v_max)
        )

    # Randbetingelser (Dirichlet)
    u_next[0] = u_left(t)     # venstre rand
    u_next[-1] = u_right(t)   # høyre rand

    return u_next


# ------------------------------------------------------------
# Gitter i tid og rom
# ------------------------------------------------------------
T = 18.0          # sluttid
nt = 200         # antall tidspunkter
a, b = -1000.0, 1000.0 # romintervall
nx = 200        # antall rompunkter

t_grid = np.linspace(0.0, T, nt)
x_grid = np.linspace(a, b, nx)

dt = t_grid[1] - t_grid[0]
dx = x_grid[1] - x_grid[0]

# ------------------------------------------------------------
# (Midleridig) testplot: viser romgitteret
# ------------------------------------------------------------
#plt.plot(x_grid, np.zeros_like(x_grid))
#plt.title("Romgitter (x-punkter)")
#plt.xlabel("x")
#plt.yticks([])
#plt.show()
#plt.close("all")

# ------------------------------------------------------------
# Vindu 2: Sett initialbetingelse, løs med Lax–Friedrichs, og animer
# Forutsetter at vindu 1 allerede er kjørt (x_grid, t_grid, dx, dt, osv.)
# ------------------------------------------------------------

# Andre Antakelser
v_max = 22.222
u_max = 0.18
p = 10

# Randbetingelser (Dirichlet)
def u_left(t):
    """Venstre randverdi u(-1000,t) = 0"""
    return 0.0;

def u_right(t):
    """Høyre randverdi u(1000,t) = 0"""
    return 0.0

# Initialbetingelse u(x,0) = u_max når x < 0, 0 når x > 0.
u0 = np.where(x_grid < 0, u_max, 0.0) # Dette var arbeid fra ChatGPT

# CFL-sjekk (pedagogisk)
cfl = (dt / dx) * np.max(np.abs(u0))
#print("max|u0| =", np.max(np.abs(u0)))
#print("CFL ~ (dt/dx)*max|u0| =", cfl)

# Lagrer løsningen for alle tider (greit for animasjon)
u = np.zeros((t_grid.size, x_grid.size))
u[0, :] = u0

# Tidsløkken
for n_step in range(1, t_grid.size):
    t_now = t_grid[n_step - 1]
    u[n_step, :] = lax_friedrichs_step(
        flux_trafikk, u[n_step - 1, :], dx, dt, t_now, u_left, u_right, u_max, v_max, p
    )

# Diagnose: sjekk at løsningen faktisk endrer seg
print("Maks endring fra t0 til t1:", np.max(np.abs(u[1, :] - u[0, :])))
print("Tiden er på: ", T, " sekund.")
print("P er: ", p)



# ------------------------------------------------------------
# Animasjon
# ------------------------------------------------------------
fig, ax = plt.subplots(figsize=(6, 4))

# Faste aksegrenser (som i eksempelet)
ax.set_xlim(x_grid[0], x_grid[-1])
ax.set_ylim(np.min(u), np.max(u))

ax.set_xlabel("x")
ax.set_ylabel("u(x,t)")
ax.set_title("Numerisk løsning med Lax–Friedrichs-metoden")

# Rød stiplet: startprofil
ax.plot(x_grid, u0, color="red", linestyle="--", label="u(x,0) (startprofil)")

# Blå kurve: løsningen som oppdateres
line, = ax.plot(x_grid, u[0, :], color="blue", label="u(x,t)")

ax.legend()

def animate(i):
    line.set_ydata(u[i, :])
    return line,

ani = animation.FuncAnimation(
    fig,
    animate,
    frames=t_grid.size,
    interval=20,
    blit=True
)

HTML(ani.to_jshtml())
plt.show()
```

### d) Bevegelsen til en enkelt bil

USIKKER

V(t) funksjonen blir en enkelt konstant av tid: V(t) = 22.22t. Det kommer fra randbetingelsen: u(-1000, t) = 0 og at V(0) = 0.
    Siden V(t) = 22.22t spiller p ingenrolle med grafen.
For å finne P(t) kan vi integrere funksjonen for fart fra 0 til t og summere initialbetingelsen, P(0) = -1000.
    Dette fyller også kravet om at V(t) = v(u(P(t), t))






```python
import numpy as np
import matplotlib.pyplot as plt

v_max = 22.22  
u_max = 0.18   
p_values = [1, 2, 5] # Ulike p verdier, men de påvirker ingenting

# Tid
t = np.linspace(0, 100, 1000)  # from 0 to 100 seconds
t[0] = 1e-6

# Farts-likningen: Konstant v(t) = 22.22t 
V = v_max * np.ones_like(t)
V[0] = 0  # at t=0 exactly, V=0

# Posisjon: integrasjon numerisk
P = np.zeros_like(t)
for i in range(1, len(t)):
    P[i] = P[i-1] + 0.5 * (V[i] + V[i-1]) * (t[i] - t[i-1])
P += -1000 # Inital

# Plotting
plt.figure(figsize=(12, 5))

# Fart-plotting
plt.subplot(1, 2, 1)
plt.plot(t, V, 'b-', linewidth=2, label=f'p = 1, 2, 5')
plt.scatter([0], [0], color='red', zorder=5, label='V(0)=0')
plt.xlabel('Time t (s)')
plt.ylabel('Velocity V(t) (m/s)')
plt.title('Velocity of the car starting at x = -1000')
plt.grid(True)
plt.legend()

# Posisjons-plot
plt.subplot(1, 2, 2)
plt.plot(t, P, 'g-', linewidth=2, label=f'p = 1, 2, 5')
plt.scatter([0], [-1000], color='red', zorder=5, label='P(0) = -1000')
plt.xlabel('Time t (s)')
plt.ylabel('Position P(t) (m)')
plt.title('Position of the car starting at x = -1000')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### e) Hastigheten til en kø
Hvis vi ser på grafen tidligere nærme så kan vi se at bilene er i bevegelsene når bølgen (fortynningsbølgen) når dem. Denne er en fortynningsbølge siden u_L (venstre siden, x < 0) er u_max og endringen av fluksen (altså den deriverte) i høyre siden er større.  Bølge beveger ikke seg mot høyre men mot venstre (Dette lager mening på grunn av reaksjonstiden til bilsjåførere i virkeligheten). Vi må ser mer på bølgen fra 0 som går mot venstre:

La u(x,t) i fortynningsbølgen for x_L(t) < x < 0, der x_L(t) er venstre grense for bølgen ved tid t.
xL(t) følger hastigheten F′(u_max)= −p * v_max * t

Dvs at bilen som akkurat nådd fortynningsbølgen (altså begynner å kjøre), er bilen ved x = x_t(t):
    K(t) = -p * v_max * t

Når er alle bilene i bevegelse? Når den siste bilen er i bevegelse. Dvs at fortynningsbølgen når x = -1000.
    -p * v_max * t = -1000
    t = 1000/p*v_max
    t = 1000/p*22.222

Ulike verdier av p gir kortere tider for at alle biler begynner å kjøre:
For p=1: t≈45.0
